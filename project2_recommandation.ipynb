```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
download_data=pd.read_csv('data2/download.csv').sort_values(by='user_id')
download_data
```

```python
user_model_pivot=pd.pivot_table(download_data,index='user_id',columns='model_id',values='created_at',aggfunc='count').fillna(0)
user_model_pivot
```

```python
model_user_pivot=pd.pivot_table(download_data,index='model_id',columns='user_id',values='created_at',aggfunc='count').fillna(0)
model_user_pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity

df_similarity=pd.DataFrame(
    data=cosine_similarity(model_user_pivot,model_user_pivot),
    index=model_user_pivot.index,
    columns=model_user_pivot.index)
df_similarity
```

```python
def recommand_item_based_CF(id):
    return df_similarity[id].sort_values(ascending=False)[1:6]

recommand_item_based_CF(64)
```

```python
from sklearn.decomposition import TruncatedSVD

SVD=TruncatedSVD(n_components=16)
matrix=SVD.fit_transform(model_user_pivot)
matrix
```

```python
import numpy as np
corr=np.corrcoef(matrix)
corr.shape
```

```python
model_id=user_model_pivot.columns
df_corr=pd.DataFrame(data=corr,index=model_id,columns=model_id)
df_corr
```

```python
def recommand_item_based_LF(id):
    return df_corr[id].sort_values(ascending=False)[1:6]

recommand_item_based_LF(64)
```
